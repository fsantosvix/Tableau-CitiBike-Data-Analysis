```python
# Import dependencies 
import pandas as pd
from pathlib import Path
```

```python
# Import Data from CSV files downloaded from the main source
df_Jan = pd.read_csv(Path("CSV_files/2023-citybike-data-NewJersey/JC-202301-citibike-tripdata.csv"))
df_Feb = pd.read_csv(Path("CSV_files/2023-citybike-data-NewJersey/JC-202302-citibike-tripdata.csv"))
df_Mar = pd.read_csv(Path("CSV_files/2023-citybike-data-NewJersey/JC-202303-citibike-tripdata.csv"))
df_Apr = pd.read_csv(Path("CSV_files/2023-citybike-data-NewJersey/JC-202304-citibike-tripdata.csv"))
df_May = pd.read_csv(Path("CSV_files/2023-citybike-data-NewJersey/JC-202305-citibike-tripdata.csv"))
df_Jun = pd.read_csv(Path("CSV_files/2023-citybike-data-NewJersey/JC-202306-citibike-tripdata.csv"))
df_Jul = pd.read_csv(Path("CSV_files/2023-citybike-data-NewJersey/JC-202307-citibike-tripdata.csv"))
df_Aug = pd.read_csv(Path("CSV_files/2023-citybike-data-NewJersey/JC-202308-citibike-tripdata.csv"))
df_Sep = pd.read_csv(Path("CSV_files/2023-citybike-data-NewJersey/JC-202309-citibike-tripdata.csv"))
df_Oct = pd.read_csv(Path("CSV_files/2023-citybike-data-NewJersey/JC-202310-citibike-tripdata.csv"))
df_Nov = pd.read_csv(Path("CSV_files/2023-citybike-data-NewJersey/JC-202311-citibike-tripdata.csv"))
df_Dec = pd.read_csv(Path("CSV_files/2023-citybike-data-NewJersey/JC-202312-citibike-tripdata.csv"))
```

```python
# Check the size of each dataset
print(f'length of Jan data: {len(df_Jan)}')
print(f'length of Feb data: {len(df_Feb)}')
print(f'length of Mar data: {len(df_Mar)}')
print(f'length of Apr data: {len(df_Apr)}')
print(f'length of May data: {len(df_May)}')
print(f'length of Jun data: {len(df_Jun)}')
print(f'length of Jul data: {len(df_Jul)}')
print(f'length of Aug data: {len(df_Aug)}')
print(f'length of Sep data: {len(df_Sep)}')
print(f'length of Oct data: {len(df_Oct)}')
print(f'length of Nov data: {len(df_Nov)}')
print(f'length of Dec data: {len(df_Dec)}')
```

```python
# Remove incomplete records and save as a reduced dataset for each month
redJan = df_Jan.dropna()
redFeb = df_Feb.dropna()
redMar = df_Mar.dropna()
redApr = df_Apr.dropna()
redMay = df_May.dropna()
redJun = df_Jun.dropna()
redJul = df_Jul.dropna()
redAug = df_Aug.dropna()
redSep = df_Sep.dropna()
redOct = df_Oct.dropna()
redNov = df_Nov.dropna()
redDec = df_Dec.dropna()
```

```python
# Check again the size of each dataset to compare with the original lengt after removing incomplete data
print(f'length of redJan data: {len(redJan)}')
print(f'length of redFeb data: {len(redFeb)}')
print(f'length of redMar data: {len(redMar)}')
print(f'length of redApr data: {len(redApr)}')
print(f'length of redMay data: {len(redMay)}')
print(f'length of redJun data: {len(redJun)}')
print(f'length of redJul data: {len(redJul)}')
print(f'length of redAug data: {len(redAug)}')
print(f'length of redSep data: {len(redSep)}')
print(f'length of redOct data: {len(redOct)}')
print(f'length of redNov data: {len(redNov)}')
print(f'length of redDec data: {len(redDec)}')
```

```python
# Export the reduced data to CSV files
redJan.to_csv('CSV_files/by_month/1.January.csv', index=False)
redFeb.to_csv('CSV_files/by_month/2.February.csv', index=False)
redMar.to_csv('CSV_files/by_month/3.March.csv', index=False)
redApr.to_csv('CSV_files/by_month/4.April.csv', index=False)
redMay.to_csv('CSV_files/by_month/5.May.csv', index=False)
redJun.to_csv('CSV_files/by_month/6.June.csv', index=False)
redJul.to_csv('CSV_files/by_month/7.July.csv', index=False)
redAug.to_csv('CSV_files/by_month/8.August.csv', index=False)
redSep.to_csv('CSV_files/by_month/9.September.csv', index=False)
redOct.to_csv('CSV_files/by_month/10.October.csv', index=False)
redNov.to_csv('CSV_files/by_month/11.November.csv', index=False)
redDec.to_csv('CSV_files/by_month/12.December.csv', index=False)
```

```python
# Combine the monthly data into a single dataset
union_df = pd.concat([redJan,redFeb,redMar,redMay, redApr,redJun,redJul,redAug,redSep,redOct,redNov,redDec])
```

```python
# Export the combined data as a CSV file to be used in Tableau
union_df.to_csv('CSV_files/Complete.csv', index=False)
```
